Interest Rates have been said to be one of the key indicators of the economy, with global investors placing heavy weightage on changes in the interest rates to manage their investments decisions. Governments have also traditionally used interest rates to manage their economy, as part of their monetary policy, with the US Federal Reserve being one example.

This project sets out to determine the impact of interest rates on key investments such as gold and shares. In addition, it also aims to uncover interesting relationships that are not traditionally taught in a finance class.

```python
# import essential libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('clean_data.csv')
```

```python
# removed the 'Year' column as we are just finding the relationship between the different factors
items = data.iloc[:, 1:]
```

```python
# calculate the correlation between the variables
corr = items.corr()
```

```python
# codes to plot the heatmap on sns
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(7, 5))
    ax = sns.heatmap(corr, mask=mask, square=True, center = 0, annot=True)
```

From the Heatmap generated, we can see that Interest Rate is negatively correlated with all major investments. In particular, it is most related to Gold, having a correlation of -0.9. 

The major investments are positively correlated with each other. Traditional knowledge tells us that as the economy grows, all major investments tend to grow, albeit at different rates.

One item that stands out is being 'untraditional' is Oil. Oil prices does not seem to have any relation with the S&P500. This is an interesting finding as one can assume that economic growth increases demand in production, which should in theory drive oil prices up. 

Based on the findings from our initial data exploration, we are interested in a more thorough investigation into 2 interesting findings:
i) The relationship between Interest Rates and Gold
ii) The lack of relationship between Oil prices and S&P500.
