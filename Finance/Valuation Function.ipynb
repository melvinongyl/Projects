Using the correlation, we are able to create a simple function that can estimate the value of a portfolio, given a certain change in interest rates

```python
# import libraries
import matplotlib.pyplot as plt 
from IPython.display import display
```

```python
# function to gather data from user
import matplotlib.pyplot as plt 
def get_data():
    in_shares = float(input('How much is in shares?'))
    in_gold = float(input('How much is in gold?'))
    in_reit = float(input('How much is in REITs?'))
    in_oil = float(input('How much is in oil?'))
    labels = ['Shares', 'Gold', 'REITs', 'Oil']
    dollars = [in_shares, in_gold, in_reit, in_oil]
    total = in_shares + in_gold + in_reit + in_oil
    fig1, ax1 = plt.subplots();    
    print('Here is the current break down of your portfolio')
    ax1.pie(dollars, labels=labels, shadow=True, startangle=90, autopct='%1.1f%%');
    display(fig1);
    plt.close()
    return dollars
```

```python
# function to determine impact of interest rates on the asset class
def interest_rate_impact(list):  
    ir_change = float(input("Based on the recent news, what is the % change of interest rate"))
    dollars = list
    dollars[0] *= (ir_change * (-0.61)/100 + 1) 
    dollars[1] *= (ir_change * (-0.9)/100 + 1) 
    dollars[2] *= (ir_change * (-0.48)/100 + 1) 
    dollars[3] *= (ir_change * (-0.39)/100 + 1) 
    portfolio_value  = dollars[0] + dollars[1] + dollars[2] + dollars[3] 
    print("Based on the interest rate changes, your new portflio is estimated to be:$", round(portfolio_value,2))
    labels = ['Shares', 'Gold', 'REITs', 'Oil']
    fig1, ax1 = plt.subplots()
    print("Here is the new breakdown of your portfolio")
    ax1.pie(dollars, labels=labels, shadow=True, startangle=90, autopct='%1.1f%%')
    display(fig1)
    plt.close()
```

```python
# combine the 2 functions into a main()
def main():
    interest_rate_impact(get_data())
```

```python
main()
```

The Valuation Function created allows the user to quickly determine the estimated value of his portfolio just by entering the estimated change in the interest rate. This is useful when the US Federal reserves announces cuts to the interest rates. Other use case would be global events that could potentially affect interest rates, in that case, the user have to use his own judgement to estimate the interest rate change. 

One limitation of the function is that it only accepts the four main asset class: Shares, REITs, Oil, and Gold. Should the user have other types of assets such as commodities or alternative investments, the function would not be able to compute. This limitation arises from the data used to calculate the correlation, whereby only a limited number of asset classes were used.

Furthermore, it should be noted that literature review tells us that the 4 asset classes are influenced not only by interest rates, but also by other factors such as foreign exchange and inflation. Those factors are not included in the Valuation Function, and would explain if the user's portolio deviates from the estimated figure. 

Despite the limitations, the Valuation Function shows us the speed at which Python can give a estimate of the portfolio value, so that the user can make decisions faster. The function also allows the user to act on the analysis made earlier, and take actions based on the insights gathered.
